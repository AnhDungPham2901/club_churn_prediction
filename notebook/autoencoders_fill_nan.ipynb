```python
import numpy as np

def matrix_factorization(R, K, steps=5000, alpha=0.0002, beta=0.02):
    """
    R: Original matrix with missing values represented as NaN
    K: Number of latent features
    steps: Number of iterations
    alpha: Learning rate
    beta: Regularization parameter
    """
    
    # Mask matrix to ignore NaN values during the update
    mask = np.isnan(R)
    
    # Fill NaN values with zeros in the original matrix
    R_filled = np.nan_to_num(R)
    
    # Initialize user and item matrices with random values
    num_users, num_items = R.shape
    U = np.random.rand(num_users, K)
    V = np.random.rand(num_items, K)
    
    # SGD to minimize error
    for step in range(steps):
        for i in range(num_users):
            for j in range(num_items):
                if not mask[i][j]:  # if not a NaN value
                    error_ij = R_filled[i][j] - np.dot(U[i, :], V[j, :].T)
                    for k in range(K):
                        U[i][k] += alpha * (2 * error_ij * V[j][k] - beta * U[i][k])
                        V[j][k] += alpha * (2 * error_ij * U[i][k] - beta * V[j][k])
    
    # Reconstructed matrix
    R_hat = np.dot(U, V.T)
    return R_hat

# Sample matrix with missing values represented as NaN
R = np.array([
    [5, 3, np.nan, 1],
    [4, np.nan, np.nan, 1],
    [1, 1, np.nan, 5],
    [1, np.nan, np.nan, 4],
    [np.nan, 1, 5, 4],
])

# Factorize and fill missing values
R_filled = matrix_factorization(R, K=2)

R_filled
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model

# Loading the raw data
df_raw = pd.read_excel('club_churn_train.xlsx')

# Preprocessing: Convert categorical columns to string type
categorical_cols = df_raw.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df_raw[col] = df_raw[col].astype(str)

# One-hot encoding for categorical columns
df_encoded = pd.get_dummies(df_raw)

# Splitting the data into train and test sets
train_data, test_data = train_test_split(df_encoded, test_size=0.2, random_state=42)

# Using only complete cases from train_data to train the autoencoder
train_data_complete = train_data.dropna()
test_data_complete = test_data.dropna()

# Scaling the data
scaler = MinMaxScaler()
train_data_scaled = scaler.fit_transform(train_data_complete)
test_data_scaled = scaler.transform(test_data_complete)

# Autoencoder architecture
input_layer = Input(shape=(train_data_scaled.shape[1],))
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)
decoded = Dense(128, activation='relu')(encoded)
output_layer = Dense(train_data_scaled.shape[1], activation='sigmoid')(decoded)

autoencoder = Model(input_layer, output_layer)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Training the autoencoder
autoencoder.fit(train_data_scaled, train_data_scaled, epochs=50, batch_size=256, shuffle=True, validation_data=(test_data_scaled, test_data_scaled))
```
